Marvel

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
marvel=pd.read_csv('Character.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment_x'].value_counts()
```

```python
good=marvel[marvel['Alignment_x']=='good']
```

```python
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False)
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power_full=good[good['Power']==100]
```

```python
good_max_power_full.shape
```

```python
good_max_power_full
```

```python
good_max_power=good_max_power_full.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color='g')
plt.show()
```
